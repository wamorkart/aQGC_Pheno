```python
import matplotlib.pyplot as plt
plt.rcParams['text.usetex'] = True
```

```python
indir = "/lcrc/project/aqgc/twamorkar/aqgc_pheno/"
# process = ["sswwjj","oswwjj", "www", "wwz", "wzjj", "wzz", "zzjj", "zzz"]
process = ["zzz", "wzz", "www"]# "wwz", "www"]
operator = "M2"

# vals = ["1e-20", "1e-12", "1e-11", "1e-10", "1e-09", "1e-08", "1e-07"]
```

```python
XS = {}
for p in process:
    XS_vals = []
    op_vals = []
    f_in = indir + "cross_section_"+p+"_aQGC_"+operator+"_SM_Ltotal_Ind5v2020v2_UFO_round3.txt"
#         print (f_in)
    with open(f_in) as f:
        lines = f.readlines()
#         print (lines)
        for l,line in enumerate(lines):
            if (l!=0):
#                 print (line.split()[2])
                op_vals.append(float(line.split()[1].replace(p+"_aQGC_"+operator+"_", "")))
                XS_vals.append(float(line.split()[2]))
#             print (line[1])
#         for line in (lines[1]):
#             print (lines[1].split()[2])
#             XS_vals.append(float(lines[1].split()[2]))
    
    XS[p] = XS_vals

    print (op_vals)
#     print (XS_vals)
    print (len(XS_vals))
    plt.plot(op_vals, XS_vals, marker="o")
#     print (p, " " , XS_vals)
# print (XS)
```
